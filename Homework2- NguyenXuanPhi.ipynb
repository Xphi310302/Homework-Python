# **TOPIC 1**
Write a program to input a floating number from keyboard and do the following things: 
-	A. The input number must have meaningful decimal part. If not, ask input again.
-	B. Separate number into two lists. The first list contains integer part and the second list contains decimal part 


```python
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
# Input number
# A
while True:
    try:
        strnumber = input('Input floating number:')
        if float(strnumber) == int(float(strnumber)):
            print('Wrong format, please reinput.')
            continue
        if str(np.round(float(strnumber),decimals = 15)) == strnumber:
            break
    except ValueError as e:
        print('Error found:', e)
        print('Please, reinput')
#B
Intstr = strnumber.split('.')[0].split('-')[-1]
Decstr = strnumber.split('.')[1]
IntList = [int(i) for i in Intstr]
DecList = [int(i) for i in Decstr]
print('Input =', strnumber)
print('IntList =', IntList)
print('DecList =', DecList)
```

Write a program to input three element : a string and two numbers. Then, do the following things: 
-	A.The string input must be in List [“GCD”, “LCM”, “FRAC”] and two number must be positive integer numbers. If not, ask input again. 
-	B.Check the string if : GCD  ==> Calculate greatest common divisor of two numbers, LCM ==> Calculate lowest common multiple of two numbers, FRAC ==> Reduce the fraction a/b 


```python
import numpy as np
from fractions import Fraction
```

```python
#A
options = ['GCD', 'LCM', 'FRAC']
while True:
    option = input("Input your option, it must be one of ['GCD', 'LCM', 'FRAC']: ")
    if option in options: break
while True:
    try:
        a = input('Please input the first positive interger:')
        b = input('Please input the second positive interger:')
        if not(float(b) == int(float(b)) and float(a) == int(float(a))) or int(a) <= 0 or int(b) <= 0:
            print('You must input postive interger, please reinput')
            continue
    except ValueError as e:
        print('Error found:', e)
        print('Please, reinput')
    else: break
if option == 'GCD':
    result = np.gcd(int(a),int(b))
if option == 'LCM':
    result = np.lcm(int(a),int(b))
if option == 'FRAC':
    result = Fraction(int(a),int(b))
print(f'Calculating {option} of {a} and {b} equal:', result)
```

Read the csv file “BigMartSales Prediction/train.csv” (Lesson 02) into a dataframe and do the following things:
-	A. Write a function to input name of statistics measurement and an attribute column and return the statistical value
-	B.Write a function to input a List of Conditions and a List of Logic Operators and return a sub dataframe after querying




```python
import pandas as pd
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
file = 'D:/Python For AI/AI/BigMartSales Prediction/Train.csv'
BigMartSalesDF = pd.read_csv(file)
BigMartSalesDF.head()
```

```python
print('Describe table of object in Dataframe')
BigMartSalesDF.describe(include=[object])
print('\nDescribe table of number in Dataframe')
BigMartSalesDF.describe(include=[np.number])
```

```python
#A
def StatisticInfo(method, feature, data = BigMartSalesDF):
    methods = {'count':lambda x: x.count(), 'mean':lambda x: x.mean(), 'std':lambda x: x.std(), 'min':lambda x: x.min(), 
    'max':lambda x: x.max(), 'unique':lambda x: x.unique(), 'mode':lambda x: x.mode(), 'median':lambda x: x.median()}
    try:
        return methods[method](data[feature])
    except:
        print(f"This column don't have {method} statistics measurement")
```

```python
Statisticlist = ['count', 'mean', 'std', 'min', 'max', 'unique', 'mode', 'median']
while True:
    print('Input statistics measurement, it should be one of those in the below list:')
    print(Statisticlist)
    method = input()
    print()
    if method in Statisticlist: break
while True:
    print("Input column's name, it should be one of those in the below list:")
    print(list(BigMartSalesDF.columns))
    feature = input()
    print()
    if feature in BigMartSalesDF.columns: break
print(f'The {method} of {feature} is:\n{StatisticInfo(method,feature)}')
```

```python
#B
def Query(conditionlist, operatorlist, data = BigMartSalesDF):
    try:
        if len(conditionlist) - len(operatorlist) != 1:
            print('Sorry, wrong input')
        else:
            operatorlist = [i.lower() for i in operatorlist]
            operator = {'and':lambda x,y:x & y, 'or':lambda x,y:x | y}
            filter = operator[operatorlist[0]](conditionlist[0],conditionlist[1])
            for i in range(1,len(operatorlist)):
                filter = operator[operatorlist[i]](conditionlist[i+1],filter)
            return data[filter]
    except:
        print('Sorry, wrong input')
```

```python
condi1 = BigMartSalesDF['Item_Weight'] < 10
condi2 = BigMartSalesDF['Item_Fat_Content'] == 'Low Fat'
condi3 = BigMartSalesDF['Item_Type'] == 'Fruits and Vegetables'
QueryDF = Query([condi1, condi2, condi3], ['AND', 'AND'])
display(QueryDF.head(8))
```
